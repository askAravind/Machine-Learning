```python
item_id = train.Item_Identifier.unique()
```

```python
dict_weight={}
for i in item_id:
    temp = (train[train['Item_Identifier']==i].Item_Weight.unique())
    for j in temp:
        if np.isnan(j)==False:
            if i not in dict_weight:
                dict_weight[i]=j
```
